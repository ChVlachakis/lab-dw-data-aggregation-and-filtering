2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
insurance_updated_df = pd.read_csv("insurance_company_cleaned_copy.csv")
insurance_updated_df
```

```python
# first we need to merge the column response to our original dataframe
merged_new_df = insurance_updated_df.merge(
    marketing_customer_analysis_df[["customer", "response"]],
    on="customer",
    how="inner"
)
merged_new_df.head(5)
```

```python
premium_value_df = (merged_new_df[merged_new_df["response"] == "Yes"].groupby(
    ["policy_type", "gender"])[["monthly_premium_auto", "customer_lifetime_value"]].mean().rename(
        columns={
            "monthly_premium_auto": "average_monthly_premium_auto",
            "customer_lifetime_value": "average_customer_lifetime_value",
        }
    )
)
premium_value_df
```

```python
# now let's add "total_claim_amount" to our analysis to find patterns.
value_to_claim_df = (merged_new_df[merged_new_df["response"] == "Yes"].groupby(
    ["policy_type", "gender"])[["monthly_premium_auto", "customer_lifetime_value", "total_claim_amount"]].mean().rename(
        columns={
            "monthly_premium_auto": "average_monthly_premium_auto",
            "customer_lifetime_value": "average_customer_lifetime_value",
            "total_claim_amount": "average_total_claim_amount"
        }
    )
)
value_to_claim_df
```

```python
value_to_claim_df["average_profit"] = value_to_claim_df["average_customer_lifetime_value"] - value_to_claim_df["average_total_claim_amount"]
value_to_claim_df
```

```python
# from the above analysis we can see that the most profitable segement for the company is the insurance tpye "special auto" and the demographic of "male", and the least profitable the same tpye but for female.
```
